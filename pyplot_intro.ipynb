
<h1><b>plypot</b></h1>
<ul>
    <li><strong>pyplot </strong>merupakan koleksi atau sebuah fungsi yang menjadikan MATPLOTLIB dapat bekerja menyerupai MATLAB</li>
        <li><strong>pyplot </strong>API secara umum less-flexible bila dibandingkan object-oriented API</li>

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

<h2><b>Membuat plotting sederhana</b></h2>

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()

#plt.show untuk merapikan lebih rapi tanpa membrikan output yang tidak dibutuhkan
```

<ul>
    <li>Pemanggilan fungsi <mark>pyplot()</mark> dapat dilakukan dengan menyertakan hanya sebuah deret bilangan <mark>list/array </mark> ataupun dua buah bilangan <mark>list/array </mark></li>
        <li>Bila pemnaggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point untuk nilai sumbu y, sedangkan pada sumbu x akan dibuatkan sebuah deret bilangan berurut dengan nilai dimulai dari nol  </li>
</ul>

```python
plt.plot([1, 3, 5, 7], [2, 4, 5, 8])
plt.show()
```

<h1>Pengaturan format pada plot</h1>
<br>
<br>
<ul>
    <li>Selain menyertakan dua buah dert bilangan sebagai parameter untuk sumbu x dan y, terdapat parameter ketiga untuk mengatur warna dan jenis plotting</li>
    <li>Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formating pada MATLAB</li>
    <li>Nilai defaultnya dalah 'b-1'</li>
    <li>Pemanggilan fungsi axis() menyertakan sebuah list <mark>[xmin, xmax, ymin, ymax]</mark></li>
    </ul>

```python
plt.plot([1, 3, 5, 7], [2, 4, 5, 8], 'ro')
#0 untuk dot , - untuk dash, * untuk bintang
plt.axis([0, 10, 0, 20])
plt.show
```

<h1>Multi plot dalam satu pemanggilan fungsi plot()</h1>

```python
t = np.arange(0., 5., 0.2,)
#nilai min, nilai max, interval nilai
t
#perbedaan linspace() pada parameter ke 3 mengatur jumlah datapoint
#arrange interval
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

<h2>Kita juga dapat melakukan multi plotting dalam satu fungsi plot()</h2>

```python
plt.plot(t, t, 'r--', t, 
         t**2, 'bs', 
         t, t**3, 'g^')
plt.show()
```

<h1>Plotting dengan keywords</h1>
<br>
<br>
<p>Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords arguments (kwargs) seperti pada <mark> dictionary </mark>dan <mark> Pandas DataFrame</mark></p>

```python
#key a = jumlah 50
#key c = random randint 0 - 50 sebanyak 50
#key d = random hanya 50 data point lalu di absolute * 100
#key b = mengadopsi nilai a + 10 dikalikan random bil 50

data = {'a' : np.arange(50),
        'c' : np.random.randint(0, 50, 50),
        'd' : np.random.randn(50)    
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
#melakukan cloning dengan scatter, parameter terakhir adalh data
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
#manmpilkan dimensi ke-3 dan ke-4
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
