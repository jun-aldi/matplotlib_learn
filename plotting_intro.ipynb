<center><h1>Matplotlib</h1></center>


matplotlib adalah salah satu module dalam python yang dapat digunakan untuk mevisualisasikan data dalam bentuk graphs

Dalama matplotlib data akan dikembangkan pada suatu figure , yang didalamnya terdiri dari satu atau lebih axes

```python
%matplotlib inline 
```

```python
import matplotlib.pyplot as plt, numpy as np
```

Membuat plotting sederhana

membuat figure yang memiliki sebuah axes 
melakukan ploting pada axes

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

fig, ax = plt.subplots() #membuat sebuah figure dan axes dengan subplots
#fig menampung object figure, axs untuk axes

ax.plot(x, y) #melakukan plotting pada data axes
```

Kita juga bisa memanfaatkan method plot() pada plypot untuk melakukan plotting sederhana

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
```

<h2>Mengenal Object Oriented Style untuk melakukan plotting</h2>

Secara eksplisit membuat figures dan axes dan memanggil methods keduannya

```python
#siapkan dataset ndengan np.linspace
x = np.linspace(0, 2, 100) #0 awal 2 akhir 100 datapoint dan intervalnya konsisten
x
```

```python
fig, ax = plt.subplots() #membuat figure dan axes

#plotting tiga varian data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label') #menyertakan x-label pada axes
ax.set_ylabel('y label') #menyertakan y-label pada axes
ax.set_title('Simple Plot') #menyertakan title pada axes
ax.legend() #menyertakan legend
```

<h2>Mengenal pyplot style untuk melakukan plotting</h2>

memanfaatkan pyplot untuk membuat figure dan axes, serta menggunakan fungsi pada pyplot untuk melakukan plotting

```python
x = np.linspace(0, 2, 100)
x
```

```python
#plotting tiga variant pada data axes
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label') #menyertakan x-label pada axes
plt.ylabel('y label') #menyertakan y-label pada axes
plt.xlabel('Simple Plot') #menyertakan tittle pada axes
plt.legend() #menyertakan legend
```
